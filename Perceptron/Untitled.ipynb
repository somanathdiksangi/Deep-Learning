```python
data=pd.read_csv("placement (2).csv")
```

```python
data
```

```python
df=data[["cgpa","iq","placement"]]
```

```python
df
```

```python
sns.scatterplot(x="cgpa",y="iq",data=df,hue="placement")
plt.show()
```

```python
x=df.iloc[:,:-1]
```

```python
x
```

```python
y=df['placement']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, X_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import Perceptron
```

```python
pr=Perceptron()
```

```python
pr.fit(X_train,y_train)
```

```python
pr.score(X_train,y_train)*100 , pr.score(X_test,y_test)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=pr)
plt.show()
```
